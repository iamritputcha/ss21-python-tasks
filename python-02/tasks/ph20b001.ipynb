```python
import numpy as np
data=hdulist[0].data
```

```python
print('Min:',np.min(data[0]))
print('Max:',np.max(data[0]))
print('Stdev:',np.std(data[0]))
```

# 6

```python
plt.rcParams['figure.figsize'] = [15,10]
```

```python
plt.imshow(data[0],cmap='gray',vmin=np.min(data[0]),vmax=np.max(data[0]))
plt.title('image')
plt.show()
```
